```python
import pandas as pd
```

```python
vine_df = pd.read_csv("../vine_table.csv", header=None, names = ["review_id","star_rating","helpful_votes","total_votes", "vine", "verified_purchase"])

vine_df.head()
vine_df.sample(n=10, random_state=1)
```

```python
#filter by total votes greater than 20
filtered_df = vine_df[vine_df['total_votes']>=20]
filtered_df.head()
filtered_df.info()
```

```python
#helpful/total >50

helpful_df = filtered_df[filtered_df['helpful_votes']/filtered_df['total_votes']>=0.5]
helpful_df.info()
```

```python
#filter for vine review only vine == Y
vine_df = helpful_df[helpful_df['vine'] == "Y"]
vine_df.info()
vine_df
```

```python
#filter for not vine, vine == N
not_vine_df = helpful_df[helpful_df['vine'] == "N"]
not_vine_df.info()
not_vine_df
```

```python
#Determine the total number of reviews, 
vine_reviews = len(vine_df)
notvine_reviews = len(not_vine_df)
#the number of 5-star reviews, and 
vine_5star = len(vine_df[vine_df['star_rating']==5])
notvine_5star =  len(not_vine_df[not_vine_df['star_rating']==5])

#the percentage of 5-star reviews for the two types of review (paid vs unpaid).
vine_per = round(vine_5star/vine_reviews*100, 1)
notvine_per = round(notvine_5star/notvine_reviews*100, 1)


summary = {"Type": ["Vine", "Not Vine"], "Total Reviews": [vine_reviews, notvine_reviews], "5 Star Reviews": [vine_5star, notvine_5star], "% 5 Star": [vine_per, notvine_per]}
summary_df = pd.DataFrame(data=summary)
summary_df
```
